```python
df = pd.read_csv("C:\\Users\\sunilverma\\Pictures\\49864_274957_bundle_archive\\google.csv",nrows = 1000)
df.shape
```

```python
df
```

```python
x = df["Rating"]
y = df["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google play store apps scattered plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r",marker = '*',s=100,alpha = 0.5,linewidths=10,
           edgecolors = "g")
plt.title("Google play store apps scattered plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r",marker = '*',s=100,alpha = 0.5,linewidths=10,
           edgecolors = "g")

plt.scatter(x,df["Installs"], c = "y",marker = 'o',s=100,alpha = 0.5,linewidths=10,
           edgecolors = "c")
plt.title("Google play store apps scattered plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
```
