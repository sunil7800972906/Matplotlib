# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "Ai","ML", "DS"]
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,width = 0.2,align = 'edge',
        color = "y",edgecolor = "m",linewidth = 5, alpha = 0.9, linestyle = "--",
       label = "Clas 1 Students")
```

```python
style.use("ggplot")
plt.figure(figsize=(19,9))
plt.bar(classes,class1_students,width = 0.6,align = 'edge',
        color = "k",edgecolor = "m",linewidth = 5, alpha = 0.9, linestyle = "--",
       label = "Clas 1 Students")
```

```python
plt.figure(figsize=(19,9))
plt.barh(classes,class1_students,align = 'edge',
        color = "k",edgecolor = "m",linewidth = 5, alpha = 0.9, linestyle = "--",
       label = "Clas 1 Students")
```

```python
plt.figure(figsize=(19,9))
plt.bar(classes,class1_students,align = 'edge',
        color = "k",edgecolor = "m",linewidth = 5, alpha = 0.9, linestyle = "--",
       label = "Clas 1 Students")
plt.title("Bar Chart of IAIP Classes", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No of Students",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(19,9))
plt.bar(classes,class1_students,width = 0.2,
        color = "b",
       label = "Clas 1 Students")
plt.bar(classes,class2_students,width = 0.2,
        color = "g",
       label = "Clas 2 Students")
plt.title("Bar Chart of IAIP Classes", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No of Students",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(19,9))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index,class1_students,width,
        color = "b",
       label = "Clas 1 Students")

plt.bar(classes_index + width,class2_students,width,
        color = "g",
       label = "Clas 2 Students")

plt.bar(classes_index + width + width,class3_students,width,
        color = "y",
       label = "Clas 3 Students")
plt.xticks(classes_index + width, classes, rotation = 20)
plt.title("Bar Chart of IAIP Classes", fontsize = 18)
plt.xlabel("Classes",fontsize = 15)
plt.ylabel("No of Students",fontsize = 15)
plt.show()
```
