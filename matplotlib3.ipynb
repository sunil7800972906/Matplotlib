# Histogram

```python
import matplotlib.pyplot as plt
import numpy  as np
import random
```

```python
ml_student_age = np.random.randint(18,45, (100))
python_student_age = np.random.randint(15,44,(100))
```

```python
print(ml_student_age)
print(python_student_age)
```

```python
plt.hist(ml_student_age,)
plt.title("ML Student age histograms")
plt.xlabel("Student age cotegory")
plt.ylabel("No of students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
from matplotlib import style
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.hist([ml_student_age,python_student_age],bins,rwidth=0.8,histtype = "bar",
         orientation = "vertical", color = ["m","y"], label = ['ML Student','Py Student'])
# plt.hist(python_student_age,bins,rwidth=0.8,histtype = "bar",
#          orientation = "vertical", color = "y", label = 'Py Student')
plt.title("ML Student age histograms")
plt.xlabel("Student age cotegory")
plt.ylabel("No of students age")
plt.legend()
plt.show()
```
